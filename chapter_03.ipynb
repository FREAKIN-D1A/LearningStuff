```python
import torch

scalar = torch.tensor(3)
print(scalar)  # Output: tensor(3)
```

```python
vector = torch.tensor([3, 9, 1])
print(vector)  # Output: tensor([3, 9, 1])
```

```python
matrix = torch.tensor([
    [16, 3, 13],
    [5, 10, 11],
    [9, 6, 7]
])
print("matrix", matrix)

tensor3d = torch.tensor([
    [[0, 7], [5, 0]],
    [[1, 2], [3, 4]],
    [[6, 5], [4, 3]],
    [[2, 1], [0, 9]]
])

print("tensor3d", tensor3d)
```

```python
import torch
import numpy as np

# Creating a tensor using a constructor
a = torch.FloatTensor(3, 2)
print(f"Tensor created using constructor:\n{a}\n")

# Creating a tensor from data (list of lists)
b = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print(f"Tensor created from data:\n{b}\n")

# Creating a tensor by converting a NumPy array
np_array = np.array([[1, 2], [3, 4]])
c = torch.from_numpy(np_array)
print(f"Tensor created by converting a NumPy array:\n{c}")
```

```python
import torch
import numpy as np

# Creating a tensor using torch.zeros
a = torch.zeros(3, 4)
print(f"Tensor created using torch.zeros:\n{a}\n")

# Using inplace operation zero_
a.zero_()
print(f"Tensor after inplace zero_ method:\n{a}\n")

# Creating a tensor from data (list of lists)
b = torch.FloatTensor([[1, 2, 3], [3, 2, 1]])
print(f"Tensor created from data:\n{b}\n")

# Creating a NumPy array
n = np.zeros(shape=(3, 2))
print(f"NumPy array:\n{n}\n")

# Converting the NumPy array to a PyTorch tensor
b = torch.tensor(n)
print(f"Tensor created by converting a NumPy array:\n{b}\n")

# Creating a NumPy array with float32 data type and converting to a tensor
n = np.zeros(shape=(3, 2), dtype=np.float32)
b = torch.tensor(n)
print(f"Tensor with float32 data type:\n{b}\n")
```
